# Inference Demo

Minimal metabolite generation example using the packaged example rules.

```python
from pathlib import Path

from grail_metabolism.model.wrapper import SimpleGenerator

rules_path = Path("grail_metabolism/resources/example_rules.txt")
rules = [line.strip() for line in rules_path.read_text().splitlines() if line.strip()]
generator = SimpleGenerator(rules)
generator.generate("CCO", top_k=3)
```
